```python
# Convert date
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
# Check missing values
df.isnull().sum()
```

```python
## Exploratory Data Analysis (EDA)
```

```python
# Total Revenue
df['total_revenue'].sum()
```

```python
##Total Quantity Sold
total_units = df["quantity_sold"].sum()
total_units
```

```python
##Revenue by Product Category
df.groupby('product_category')['total_revenue'].sum().sort_values(ascending=False)
```

```python
## Revenue by Region
region_analysis = df.groupby("customer_region").agg({
    "total_revenue": "sum",
    "quantity_sold": "sum",
    "rating": "mean"
})
region_analysis
```

```python
## Best Payment Method
df.groupby('payment_method')['total_revenue'].sum()
```

```python
##Average Rating by Category
df.groupby('product_category')['rating'].mean().round(2)
```

```python
## Discount Impact Analysis
df[["discount_percent", "quantity_sold", "total_revenue"]].corr()
```

```python
## Time Analysis
df['month'] = df['order_date'].dt.month
df.groupby('month')['total_revenue'].sum()
```

```python
plt.figure()
df.groupby("product_category")["total_revenue"].sum().plot(kind="bar")
plt.title("Total Revenue by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
##Top Performing Orders
df.sort_values("total_revenue", ascending=False).head()
```

```python
##High Review but Low Rating Products
df.sort_values(["review_count","rating"], ascending=[False, True]).head()
```

```python
# Revenue by Region
plt.figure()
df.groupby("customer_region")["total_revenue"].sum().plot(kind="barh")
plt.title("Total Revenue by Customer Region")
plt.xlabel("Customer Region")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
# Revenue Over Time
plt.figure()
df.sort_values("order_date").groupby("order_date")["total_revenue"].sum().plot(kind="line")
plt.title("Revenue Over Time")
plt.xlabel("Order Date")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()
```
